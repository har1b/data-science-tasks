```python
cof=pd.read_csv("Costomer+OrderForm.csv")
```

```python
cof
```

```python
cof_gd=pd.get_dummies(cof)
```

```python
cof_gd
```

```python
cof_gd.sum()
```

```python
cof_new=cof_gd[{"Phillippines_Defective","Indonesia_Defective","Malta_Defective","India_Defective"}]
```

```python
stats.f_oneway(cof_new["Indonesia_Defective"],cof_new["Malta_Defective"],cof_new["Phillippines_Defective"],cof_new["India_Defective"],)
```

```python
# pvalue=0.2776780
```

```python
alpha= 0.05
p=0.2776780955705948
```

```python
print('significance=%.3f, p=%.3f' % (alpha, p))
if p <= alpha:
	print('Dependent (reject H0)')
else:
	print('Independent (fail to reject H0)')
```
